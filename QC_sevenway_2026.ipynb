## Turn xlsx into each month xlsx

```python
import pandas as pd

files = {
    "2025-12": "空氣品質小時值_桃園市_桃園站 (2025-12).csv",
    "2026-01": "空氣品質小時值_桃園市_桃園站 (2026-01).csv",
    "2026-02": "空氣品質小時值_桃園市_桃園站 (2026-02).csv"
}

target_cols = ["SO2", "CO", "O3", "PM10", "NO2", "PM2.5"]

all_data = []

for month, file in files.items():
    df = pd.read_csv(file, encoding="utf-8-sig")

    df["monitordate"] = pd.to_datetime(df["monitordate"])
    df["itemengname"] = df["itemengname"].astype(str).str.strip()
    df["concentration"] = pd.to_numeric(df["concentration"], errors="coerce")

    wide_df = df.pivot_table(
        index="monitordate",
        columns="itemengname",
        values="concentration",
        aggfunc="mean"
    ).reset_index()

    wide_df.columns.name = None

    wide_df = wide_df[["monitordate"] + target_cols]
    wide_df["month"] = month

    output_file = f"air_quality_{month.replace('-', '')}.xlsx"
    wide_df.to_excel(output_file, index=False)

    all_data.append(wide_df)

combined_df = pd.concat(all_data, ignore_index=True)


summary = combined_df.groupby("month")[target_cols].agg(["mean", "max", "std"])

print(summary)


pm25_summary = combined_df.groupby("month")["PM2.5"].agg(["mean", "max", "std"])

print("PM2.5 monthly summary:")
print(pm25_summary)
```

### 2. 查核表

```python
df = pd.read_excel("air_quality_202601.xlsx")

df["monitordate"] = pd.to_datetime(df["monitordate"])
df_pm25 = df.dropna(subset=["PM2.5"]).copy()

weekday_map = {
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday"
}

weekday_order = [
    "Monday", "Tuesday", "Wednesday", "Thursday",
    "Friday", "Saturday", "Sunday"
]

df_pm25["weekday"] = df_pm25["monitordate"].dt.weekday.map(weekday_map)

bins = [0, 15.4, 35.4, 54.4, 150.4, float("inf")]

labels = [
    "Good (0.0-15.4)",
    "Moderate (15.5-35.4)",
    "Unhealthy for Sensitive Groups (35.5-54.4)",
    "Unhealthy (54.5-150.4)",
    "Very High (>150.4)"
]

df_pm25["PM25_range"] = pd.cut(
    df_pm25["PM2.5"],
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=True
)

check_sheet = pd.crosstab(
    df_pm25["PM25_range"],
    df_pm25["weekday"]
)

check_sheet = check_sheet.reindex(
    index=labels,
    columns=weekday_order,
    fill_value=0
)

check_sheet["Total"] = check_sheet.sum(axis=1)

print(check_sheet)
```

### 3. 層別法

```python
df = pd.read_excel("air_quality_202601.xlsx")
df["monitordate"] = pd.to_datetime(df["monitordate"])
df["hour"] = df["monitordate"].dt.hour

def assign_period(hour):
    if 0 <= hour <= 5:
        return "Late night"
    elif 6 <= hour <= 11:
        return "Morning"
    elif 12 <= hour <= 17:
        return "Afternoon"
    else:
        return "Evening"

df["period"] = df["hour"].apply(assign_period)

period_summary = df.groupby("period")["PM2.5"].agg(["count", "mean", "max", "std"])

print(period_summary)
```

```python
import pandas as pd

df = pd.read_excel("air_quality_202601.xlsx")

df["monitordate"] = pd.to_datetime(df["monitordate"])
df["hour"] = df["monitordate"].dt.hour

def assign_period(hour):
    if 0 <= hour <= 5:
        return "Late night"
    elif 6 <= hour <= 11:
        return "Morning"
    elif 12 <= hour <= 17:
        return "Afternoon"
    else:
        return "Evening"

df["period"] = df["hour"].apply(assign_period)

df_pm25 = df.dropna(subset=["PM2.5"]).copy()

threshold = 35.4
df_pm25["is_high_pm25"] = df_pm25["PM2.5"] > threshold

high_summary = df_pm25.groupby("period")["is_high_pm25"].agg(
    high_count="sum",
    total_count="count"
)

high_summary["high_ratio"] = high_summary["high_count"] / high_summary["total_count"]

print(high_summary)
```

```python
period_high_distribution = (
    df_pm25[df_pm25["is_high_pm25"]]
    .groupby("period")
    .size()
    .reset_index(name="high_count")
)

period_high_distribution["percentage"] = (
    period_high_distribution["high_count"] 
    / period_high_distribution["high_count"].sum() * 100
)

print(period_high_distribution)
```

### 5. 柏拉圖(觀察 PM2.5 高濃度事件發生時，哪些污染物也達到 AQI 普通以上或更高等級。)-PM2.5 高濃度事件之共同偏高污染物柏拉圖Q3

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("air_quality_202601.xlsx")

pollutants = ["CO", "SO2", "NO2", "O3"]

pm25_threshold = 35.4
high_pm25_df = df[df["PM2.5"] > pm25_threshold].copy()

print("PM2.5 高濃度事件筆數：", len(high_pm25_df))

pollutant_thresholds = {}

for pollutant in pollutants:
    pollutant_thresholds[pollutant] = df[pollutant].quantile(0.75)

print("各污染物偏高門檻：")
print(pollutant_thresholds)

def classify_event(row):
    high_pollutants = []

    for pollutant in pollutants:
        if row[pollutant] > pollutant_thresholds[pollutant]:
            high_pollutants.append(pollutant)

    if len(high_pollutants) == 0:
        return "No co-pollutant high"
    else:
        return " + ".join(high_pollutants)

high_pm25_df["event_type"] = high_pm25_df.apply(classify_event, axis=1)

pareto = (
    high_pm25_df["event_type"]
    .value_counts()
    .reset_index()
)

pareto.columns = ["category", "count"]

pareto["percentage"] = pareto["count"] / pareto["count"].sum() * 100
pareto["cumulative_percentage"] = pareto["percentage"].cumsum()

print(pareto)

fig, ax1 = plt.subplots(figsize=(8, 5))

ax1.bar(pareto["category"], pareto["count"])
ax1.set_xlabel("Event Type")
ax1.set_ylabel("Count")
ax1.set_title("Pareto Chart of PM2.5 High Concentration Event Types")
ax1.set_ylim(0, pareto["count"].sum() * 1.1)

ax2 = ax1.twinx()
ax2.plot(
    pareto["category"],
    pareto["cumulative_percentage"],
    marker="o",
    color="red",
    linestyle="--"
)

ax2.set_ylabel("Cumulative Percentage (%)")
ax2.set_ylim(0, 110)

for i, value in enumerate(pareto["cumulative_percentage"]):
    ax2.text(
        i,
        value + 3,
        f"{value:.1f}%",
        ha="center",
        color="red"
    )

plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("air_quality_202601.xlsx")

pollutants = ["CO", "SO2", "NO2", "O3"]

# PM2.5 高濃度事件：仍用 AQI 對敏感族群不健康門檻
pm25_threshold = 35.4
high_pm25_df = df[df["PM2.5"] > pm25_threshold].copy()

print("PM2.5 高濃度事件筆數：", len(high_pm25_df))

# 其他污染物偏高門檻：使用該月第 75 百分位數
pollutant_thresholds = {}

for pollutant in pollutants:
    pollutant_thresholds[pollutant] = df[pollutant].quantile(0.75)

print("各污染物偏高門檻：")
print(pollutant_thresholds)


co_high_counts = {}

for pollutant, threshold in pollutant_thresholds.items():
    co_high_counts[pollutant] = (high_pm25_df[pollutant] > threshold).sum()

pareto = pd.DataFrame({
    "pollutant": list(co_high_counts.keys()),
    "count": list(co_high_counts.values())
})

pareto = pareto.sort_values("count", ascending=False)

# 只保留有共同偏高的污染物
pareto = pareto[pareto["count"] > 0].copy()

pareto["percentage"] = pareto["count"] / pareto["count"].sum() * 100
pareto["cumulative_count"] = pareto["count"].cumsum()
pareto["cumulative_percentage"] = pareto["percentage"].cumsum()

print(pareto)

fig, ax1 = plt.subplots(figsize=(7, 5))

ax1.bar(pareto["pollutant"], pareto["count"])
ax1.set_xlabel("Pollutant")
ax1.set_ylabel("Co-occurrence Count")
ax1.set_title("Pareto Chart of Co-occurring Pollutants during High PM2.5 Events")

# 左右兩邊 y 軸使用一樣的 scale
y_max = pareto["cumulative_count"].max() + 1
total_count = pareto["count"].sum()
ax1.set_ylim(0, total_count * 1.1)

ax2 = ax1.twinx()
ax2.plot(
    pareto["pollutant"],
    pareto["cumulative_percentage"],
    marker="o",
    color="red",
    linestyle="--"
)

ax2.set_ylabel("Cumulative Percentage (%)")
ax2.set_ylim(0, 110)

# 標示累積百分比
for i, value in enumerate(pareto["cumulative_percentage"]):
    ax2.text(
        i,
        value + 3,
        f"{value:.1f}%",
        ha="center",
        color="red"
    )


plt.tight_layout()
plt.show()
```
